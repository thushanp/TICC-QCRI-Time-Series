```python
import numpy as np
import matplotlib.pyplot as plt
import csv

with open("freq1clustdata1.csv") as f:
    reader = csv.reader(f)
    next(reader) # skip header
    data = [r for r in reader]
    
# testing


def column(matrix, i):
    return [row[i] for row in matrix]



x = column(data, 0)
x = [int(i) for i in x]
howmanyclust = len(set(x))

print(str(howmanyclust) + "   <-- this many clusters")

y = column(data, 1)

c = ['w','g','r','c', 'y', 'm']

y = [float(i) for i in y]


z = np.arange(1, 1440, 1.)


print(z)

```

```python
t = y
s = z

plt.plot(s, t)
```

```python
for i in range(1,1439):
    p = plt.axvspan(i, i+1, facecolor=c[x[i]], alpha=0.5)

plt.show()
```

```python
with open("twistclust1.csv") as f:
    reader = csv.reader(f)
    next(reader) # skip header
    data = [r for r in reader]


def column(matrix, i):
    return [row[i] for row in matrix]



x = column(data, 0)
x = [int(i) for i in x]
howmanyclust = len(set(x))

print(str(howmanyclust) + "   <-- this many clusters")

y = column(data, 1)

c = ['w','g','r','c', 'y', 'm', 'k', 'b', '#8F00FF', '#29AB87']

y = [float(i) for i in y]


z = np.arange(1, 2100, 1.)


print(z)
t = y
s = z
plt.figure(figsize=(20,10))
plt.plot(s, t)

for i in range(1,2095):
    p = plt.axvspan(i, i+1, facecolor=c[x[i]], alpha=0.5)

plt.show()

#BIC: 330
```
