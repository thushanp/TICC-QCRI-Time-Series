```python
import numpy as np
import matplotlib.pyplot as plt
import csv

with open("freqlong1clustdata4.csv") as f:
    reader = csv.reader(f)
    next(reader) # skip header
    data = [r for r in reader]


def column(matrix, i):
    return [row[i] for row in matrix]



x = column(data, 0)
x = [int(i) for i in x]
howmanyclust = len(set(x))

print(str(howmanyclust) + "   <-- this many clusters")

y = column(data, 1)

c = ['w','g','r','c', 'y', 'm', 'k', 'b', '#8F00FF', '#29AB87']

y = [float(i) for i in y]


z = np.arange(1, 2522, 1.)


print(z)
t = y
s = z

plt.plot(s, t)

mini = 0

for i in range(1,2520):
    nextcol = c[x[i+1]]
    cur = c[x[i]]
    if nextcol == cur:
        if mini == 0:
            mini = i
        continue
    else:
        p = plt.axvspan(mini, i, facecolor=c[x[i]], alpha=0.5)
        mini = i+1

plt.show()
```
